```python
# -------------------------------
# STEP 7: Sales by Region
# -------------------------------
# Shows which regions contribute the most to sales
plt.figure(figsize=(7,5))
sns.barplot(x="Region", y="Sales", data=df, estimator=sum, palette="magma")
plt.title("Sales by Region")
plt.show()
```

```python
# -------------------------------
# STEP 8: Discount vs Profit
# -------------------------------
# Scatterplot to see how discounts affect profit
plt.figure(figsize=(7,5))
sns.scatterplot(x="Discount", y="Profit", data=df, color="red")
plt.title("Impact of Discount on Profit")
plt.show()
```

```python
# -------------------------------
# STEP 9: Sales by Customer Segment
# -------------------------------
# Identifies which customer segment (Consumer/Corporate/Home Office) spends most
plt.figure(figsize=(7,5))
sns.barplot(x="Segment", y="Sales", data=df, estimator=sum, palette="pastel")
plt.title("Sales by Customer Segment")
plt.show()
```

```python
# -------------------------------
# STEP 10: Orders by Shipping Mode
# -------------------------------
# Shows which shipping method is most frequently used
plt.figure(figsize=(7,5))
sns.countplot(x="Ship Mode", data=df, palette="Set2")
plt.title("Orders by Shipping Mode")
plt.show()
```
